### Genrator
A generator is a function that uses yield to return values one at a time and automatically implements the iterator protocol.


```python
def square(n):
    for i in range(3):
        yield i**2 #yield keywords provides to be some perticular memroy allocation
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
a =square(3)
a
```

```python
next(a)
```

```python
def my_genrator():
    yield 1
    yield 2
    yield 3
    
```

```python
gen =my_genrator()
gen
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```

### Prectical Example : Reading Larg file


```python
### Pricatal Reading larg file

def read_larg_file(file_path):
    with open(file_path,'r') as file:
        for line in file: # line will print line by line word execting
            yield line
 
```

```python
file_path ='larg_file.txt'

for line  in read_larg_file(file_path):
    print(line.strip()) # here Strip bassically to do remove extra space and line
```
