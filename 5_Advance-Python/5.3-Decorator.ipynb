### Decorator 

A decorator is a higher-order function that takes another function as input and returns a modified version of that function without altering its source code.

```python
# Function
# Decoretors
# Cluster
```

```python
# Function  copy

def welcome():
    return 'Welcome to  Python Advance course'

welcome()    
```

```python
wl1=welcome
print(wl1())
del welcome
print(wl1())
```

```python
    # Decorator

def reapeat(n):
    def decorator(fun):  
        def warper(*arg,**kwarg):   # arg = collocting tupple  # **kwarg store dicsionary
            for _ in range(n):
                fun(*arg,**kwarg)
        return warper
    return decorator        
```
