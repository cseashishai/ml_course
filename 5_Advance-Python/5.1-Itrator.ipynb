### Itrator 
An iterator is an object in Python that returns data one element at a time and remembers its current position.

It allows you to go through (iterate over) a sequence step-by-step.

```python
my_list=[1,2,3,4,5,6]
for i in my_list:
    print (i)
```

```python
type(my_list)
```

```python
# Iterator 

iterator =iter(my_list) #when you call iterator to use iter
print(type(iterator))
```
