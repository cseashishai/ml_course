### Sql and Sqllite
SQL (Structured Query Language) is a standard language for managing and manipulating relational databases.
SQLite is a self-contained, serverless, and zero-configuration database engine that is widely used for embedded database systems.
In this lesson, we will cover the basics of SQL and SQLite, including creating databases, tables, and performing various SQL operations.

```python
import sqlite3
```

```python
## Connect sql lite Database 
connection = sqlite3.connect("example.db") # here 'example.db' foleder creting and connecting it
connection
```

```python
cursor = connection.cursor()
```

```python
# Creat a tabal
cursor.execute('''
Create Table if not Exists employs(
   id Integer Primary key,
   name Text Not Null,
   age Integer,
   department text 
  )
''')

## Commit the changes

connection.commit()
```

```python
# Creat a tabal
cursor.execute('''
              
Select * from employs
  
''')

```

```python
cursor.execute('''
INSERT INTO employs(name,age,department)
VALUES ('Ashish',20,'Beginning ML Engineer')
''')

cursor.execute('''
INSERT INTO employs(name,age,department)
VALUES ('Aman',22,'Accountant')
''')
cursor.execute('''
INSERT INTO employs(name,age,department)
VALUES ('Jay',23,'Financer')
''')

connection.commit()
```

```python
#Qury from the data to tabel 
cursor.execute('Select * FROM employs')
row =cursor.fetchall()

# print  the qury data
for row in row:
    print(row)
    
```

```python
# update the data in tabel 

cursor.execute('''
UPDATE employs
set age=22
where name ="Ashish"
''')

connection.commit()  
```

```python
#Qury from the data to tabel 
cursor.execute('Select * FROM employs')
row =cursor.fetchone() # You may be use to the many row or coloumns using "Fetchall ,one,any"

# print  the qury data
for row in row:
    print(row)
```

```python
# Delete the data from the tabel 
# Jay data rmoving
cursor.execute('''
DELETE from employs 
               where name ='Jay'   
''')

connection.commit()
```
