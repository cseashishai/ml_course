```python
## CSV fike creating
import csv

with open('Ashu.csv',mode='w',newline='') as file:
    writer =csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['ashu',32])

with open('Ashu.csv',mode='r') as file:
    read=csv.reader(file)
    for row in read:
        print(row)
```
