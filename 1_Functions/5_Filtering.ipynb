The Filter() Function in python


```python
def even(num):
    if num%2==0:
        return True
    
```

```python
# even(24)
```

```python
lst =[2,5,4,6,7,8,3,6,9,10]

list(filter(even,lst))
```

```python
# filter by lamda function

number =[1,2,3,4,5,6,7,8]
grater_tahn_five =list(filter(lambda x:x>5,number))
print(grater_tahn_five )
```
