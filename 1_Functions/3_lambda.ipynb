```python
# when i use lambda function

addition = lambda a,b:a+b
type(addition)
print(addition(5,6))
```

```python
def even_num(num):
    if num%2==0:
        return True
even_num(50)
```

```python
# lambada funcation 
even = lambda num:num%2==0
even(12)
```

```python
# map 

number = [1,2,3,4,5]
def square(number):
    return number**2

square(5)
```

```python
map(lambda x:x**2,number)
list(map(lambda x:x**2,number))


# squar = lambda value:value**2
# # square(5)
# print(square(5))
```
