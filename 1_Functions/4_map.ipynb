Map()

```python
def data(x):
    return x*x

data(10)    
```

```python
lit =[1,2,3,4,5,6,4,5]
list(map(data,lit))
```

```python
# when i using lambda function

list(map(lambda x:x*x,lit))
```

```python
# when i usnig multiple fuction
num1= [1,2,3]
num2= [4,5,6]

add =list(map(lambda x,y:x+y,num1,num2))
print(add)
```

```python
# map to convert str list to int

str_num =('1','2','3')
number=list(map(int,str_num))

print(number)
```

```python
words =("apple","banana","papaya","balck berry")

upper_Word =list(map(str.upper,words))

print(upper_Word)
```

```python
def get_name(person):
    return  person['name']

people =[
    {'name':'ashu','age':19},
    {'name':'aman','age':20},
    {'name':'boy','age':30}
]

list(map(get_name,people))
```
