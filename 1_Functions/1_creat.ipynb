```python
##keyword argument

def key_values(**kaal):
    for key,value in kaal.items():
        print(f"{key}:{value}")
```

```python
key_values(name ="ashu",age=10, countery = "india")
```

```python
# diffrece between arg and kwargs

def functions_two(*arg,**kwarg):
    for point in arg:
        print(f"Function are arg :{point}")

    for key,value in kwarg.items():
        print(f"{key}:{value}")    
```

```python
functions_two(1,2,3,4,name ="ashu",countery ="India",age =10)
```
