```python
# Define a class of instance

class Dog:

    def __init__(self,name,age):
        self.name =name
        self.age=age

    def bark(self):
        print(f"{self.name} {self.age}say woof")    

dog1 =Dog("Puppy",3)
dog1.bark()
dog2 =Dog("doggy",4)
dog2.bark()
 
```

```python
### MODELING THE BANK ACCOUNT
# Define a class  for a  bank account

class Banckaccount:
    def __init__(self,owanr,balance=0):
        self.owanr =owanr
        self.balance=balance

    def deposit(self,ammount):
        self.balance+=ammount
        print(f"{ammount} is deposit . New balance is {self.balance}")

    def withdraw(self,ammount):
        if ammount>self.balance:
            print("Influance Fund")

        else:
            self.balance-=ammount
            print(f"{ammount} is the withdraw is {self.balance}")

    def get_balance(self):
        return self.balance     

account =Banckaccount("ashu",5000)
print(account.balance)
```
