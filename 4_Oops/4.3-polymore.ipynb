### Polymorphism :

Polymorphism is the ability in object-oriented programming for different objects to respond to the same method name in their own specific way.

### Method Overiding
 Method overriding is a feature in OOP where a child class provides its own mplementation 
 of a method that is already defined in its parent class.


```python
## Base Class
class Animal:
    def Animal(self):
        return "sound of animal"
    
# creat class_1
class Dog(Animal):
    def speak(self):
        return "woof !"
       
# ceart class
class Cat(Animal):
    def speak(self):
        return "mewo! Ghap Ghap Ghap"

# Function demonstrates polymorephysm

def animal_speak(animal):
    print(animal.speak())



huggy = Dog()
cat =Cat()
print(cat.speak())
print(huggy.speak())   

animal_speak(huggy)
animal_speak(cat)
```

```python
# Polymorphism With Function and method
# Base 1

class Shape():
    def area(self):
        return "The area of figure"
    
# Dreived class 1

class Ractangle(Shape):
    def __init__(self,width,lenth):
        self.width =width
        self.lenth =lenth

    def area(self):
        return self.width*self.lenth

# Drive 2 Class

class Circule(Shape):
    def __init__(self,redius):
         self.redius=redius


    def area(self):
        return 3.14*self.redius*self.redius # area is pi* r^2

# Function  That demonstrates polymorphysum


def print_area(shape):
    print(f"The ara is {shape.area()}")


Ractangle1 =Ractangle(4,5)
Circul =Circule( 3)

print_area(Ractangle1)
print_area(Circul)    
```

### Polymorephysm with Abstact Base class (ABC)
Polymorphism with Abstract Base Class means defining a common interface (method structure) in a base class and allowing multiple child classes to implement that method in their own way.


```python
from abc import ABC,abstractmethod

# Define Abstract Classs

class Vehical(ABC):
    
    @abstractmethod #Forces child class to implement method
    def start_engine(self):
            pass
    
# Drive 1_class

class Car(Vehical):
      def start_engine(self):
            return "Car engine started " 

#Drive class_2

class Motercycle(Vehical):
      def start_engine(self):
            return "Motercyle is Start" 
       
# function demonstaration

def start_vehical(vehica):
      print(vehica.start_engine())

# Creating object of class (Car & Motercycle) 

car =Car()
mootercyle =Motercycle()

start_vehical(car)
```
