### Magic Method

Magic methods are special predefined methods in Python that allow objects to behave like built-in types by customizing operations such as initialization, printing, arithmetic, comparison, etc.

### Magic Methods

Magic methods (also called dunder methods, short for double underscore) are special methods in Python that start and end with double underscores, like:

init : Inislize a new instance of a class

str : Return the strig of object repersentaion

len : Return the lenth of object

repr : is use to returne official string repersentaion of object

getitem : Get item from the contenor

setitem : Set item in contaenor


They let you define how your objects behave with built-in Python operations such as addition, printing, comparison, length calculation, and more.

```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age

    def __str__(self):
        return f"{self.name},{self.age} year old"    
    
    def __repr__(self):
        return f"Person (name = {self.name},age = {self.age})"
    
person =Person("Ashu",25)
print(person) 
print(repr(person))
print(str(person))
```

### Operator Overlaoding

```python
class Vector:
    def __init__(self,x,y):
        self.x =x 
        self.y =y

    def __add__(self,finde):
        return Vector(self.x+finde.x,self.y+finde.y) 
    
    def __sub__(self,finde):
        return Vector(self.x-finde.x,self.y-finde.y)
    
    def __mul__(self,finde):
        return Vector(self.x*finde.x,self.y*finde.y)
    
    def __eq__(self,finde):
        return Vector(self.x == finde.x,self.y == finde.y)
    
    def __repr__(self):
        return f" Vector {self.x}, {self.y}"
    
v1 =Vector(2,3)
v2 =Vector(4,5)

print(v1+v2)
# print(v1-v2)
# print(v1*v2)
```

```python
### Overloading Operator For complex number

"""  Easy to understan to 
Real part uses:    real * real
                   imag * imag

Imaginary part uses:  real * imag
                      imag * real              

"""

class Complexnumber:
    def __init__(self,real,imag):
        self.real =real
        self.imag =imag

    def __add__(self,other):
        return Complexnumber(self.real+other.real,self.imag+other.imag) # Formula ::(a+bi)+(c+di)=(a+c)+(b+d)i

    def __sub__(self, other):
        return Complexnumber(self.real-other.real,self.imag-other.imag) #Formula :(a+bi)−(c+di)=(a−c)+(b−d)i

    def __mul__(self, other):
        real_part =  self.real * other.real - self.imag * other.imag 
        imag_part =  self.real * other.imag + self.imag * other.real 
        return Complexnumber(real_part,imag_part)
    
    def __truediv__(self, other):
        dominator = other.real**2 +other.imag**2  # Formula :c+di / a+bi = (ac+bd)+(bc−ad)i /​ c^2 + d^2

        real_part =  (self.real * other.real - self.real * other.imag) / dominator
        imag_part =  (self.real * other.imag + self.imag * other.real  ) / dominator
        return Complexnumber(real_part,imag_part)


    def __eq__(self, other):
        return self.real == other.real and self.imag == other.imag #other mean a1 +a2 here other is a2
    
    def __repr__(self):
        return f"{self.real},{self.imag}"
    
c1 =Complexnumber(2, 3)
c2 =Complexnumber(1, 4)
print(c1+c2)    
print(c1-c2)
print(c1*c2)
print(c1/c2)
print(c1==c2)
```
