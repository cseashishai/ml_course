Inheritance : 
              Inheritance is the mechanism in object-oriented programming where one class acquires the properties and behaviors (methods and attributes) of another class.

```python
# Inheritance (Single inheritance)
# Psrent clsass

class Car:
    def __init__(self,window,engeine,eginetype):
        self.window =window
        self.engeine =engeine
        self.eginetype=eginetype

    def drive(self):
        print(f"The person drive with self {self.eginetype}")
```

```python
car1=Car(4,5,"ASHISH.engine")
car1.drive()    
```

```python
# Now i just inherit it

class Tesla(Car): #here is tesla is child class / Car is parent class
    def __init__(self, window, engeine, eginetype,selfdriving):
        super().__init__( window, engeine, eginetype)# when Only 1 call (super) of prent class to call to use
        self.selfdriving=selfdriving

    def tesla_test(self):
        print(f"Tesla  Supports self driving {self.selfdriving} {self.window}")    
```

```python
tesla =Tesla(4,1,"Ashish",True)
tesla.tesla_test()
```

```python
tesla.drive() #here i call  (drive) from to  parent class to call
```

```python
# Multiple Inheritance
# Base class_1

class Animal:
    def __init__(self,name):
        self.name=name

    def speak(self):
        print('The multiple Animal are Multiple language to speak')    

# Base Class_2

class Pet:
    def __init__(self,Owner):
        self.Owner =Owner

# Drived Class 

class Dog(Animal,Pet):
    def __init__(self,name,Owner):
        Animal.__init__(self,name)
        Pet.__init__(self,Owner)

    def speak(self):
        return f"{self.name} Say woof"    

dog =Dog("Buddy","Ashish")
print(dog.speak())
print(f"Dog Owner  {dog.Owner}")
```
