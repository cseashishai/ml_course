### Encapsulation

Encapsulation is an Object-Oriented Programming (OOP) concept that means wrapping data (variables) and methods (functions) together into a single unit (class) and restricting direct access to some of the object's components.

```python
# Encapsulaation Gatter and Setter
### Public,Private and Protected varible

class Person:
    def __init__(self,name,age):

        self.name =name  #Public varible
        self.age =age #Public Variable

def get_name(person):
        return person.name    

person =Person("Ashu",34)
get_name(person)

```

```python
# Encapsulaation Gatter and Setter
### Protected varible

class Person:
    def __init__(self,name,age,gender):

        self._name =name  # protected (_name) is here it protected Variable
        self._age =age #protected (_age) is here it protected Variable
        self.gender =gender

class Employ(Person):
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)        


employ =Employ("ashu",19,"male")
print(employ._name)
```

```python
## Encapsulation Getter and setter

class Person:
    def __init__(self,name,age):

        self.__name =name  # private (__name) is here it private Variable
        self.__age =age 

# Getter method for name
    def get_name(self):
        return self.__name
    
    # getter method
    def get_age(self):
        return self.__age
    
# Setter Method for name
    def change_value(self,age):
        if age >0:
            self.__age =age
        else:   
            print("Age can not be Nagative.")

person =Person("Ashu",12)

# Access and modifying private varible and Getter and setter

print(person.get_name())
print(person.get_age())

person.change_value(20)
print(person.get_age())

person.change_value(-5)
```
