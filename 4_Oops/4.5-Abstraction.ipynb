### Abstraction
Abstraction is the process of hiding unnecessary details and showing only the essential features of something, so it’s easier to understand, use, or work with.

```python
from abc import ABC,abstractmethod

## Abstract Class
class vehical:
    def drive(self):
        print("The Vehical use to drive ")

    @abstractmethod
    def start_engine(self):
        pass

class Car(vehical):
    def start_engine(self):
        print("car engine start")

def operate_vehical(vehicale):
        vehicale.start_engine()
        vehicale.drive() 

car =Car()
operate_vehical(car)

```
