```python
# Perticuller function name that be mostly to use

arr2 =np.array([[1,2,3,4,6,7,8],[8,9,10,11,12,13,14]])

print("Array :\n",arr2)
print("Number of Dimansion :",arr2.ndim)
print("The Data type :",arr2.dtype)
print("Size of array ",arr2.size)
print("Item size is (In Byte) :",arr2.itemsize)
```

```python
# Numpy Vector operation
arr1 =np.array([[1,2,3,4,5]])
arr2 =np.array([[8,9,10,11,12]])

# Element wise addition(
print("Addition",arr1+arr2)


# Element wise subtraction
print("subtraction",arr1-arr2)

# Element wise multiplication(
print("multiplication",arr1*arr2)

# Element wise divison
print("divison",arr1/arr2)
```

```python
# Universal Function
arr1 =np.array([[1,2,3,4,5]])

# squre root
print(np.square(arr1))

## Exponesial 
print(np.exp(arr1)) #  Exponesial  is e^n  its use to ML

# use sin()
print(np.sin(arr1)) # here is sin() perform to the sine() values

# Log(ln) natrural
print(np.log(arr1))
```

```python
# Statical concepts
#  Normalization 

# to have a mean of 0 and standard divison of 1

data =np.array([1,2,3,4,5])

# Calculate the standard division 
mean =np.mean(data)
std_dev =np.std(data)

# normalizaition the data

normlized_data =(data-mean)/std_dev
print("normlized_data :" , normlized_data)
```

```python
# Finding the mean median and div

data =np.array([1,2,3,4,5])

# finding mean
mean =np.mean(data)
print(mean)

# Meadian

median =np.median(data)
print( "Median is :",median)

# Division
Div =np.std(data) 
print("Divison are :",Div)

# Variance 
var =np.var(data)
print("variance is that :",var)
```
