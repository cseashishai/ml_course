```python
# Data agrigating and grupping

group_mean =df.groupby('Loan_ID')['CoapplicantIncome'].mean()
print(group_mean)
```

```python
# Merging and joing data frame
df1 =pd.DataFrame({"key":['A','B','C'], 'Value1':[1,2,3]})
df2 =pd.DataFrame({"key":['A','B','D'], 'Value2':[4,5,6]})
```

```python
print(df1)
print(df2)
```

```python
# Merge data frame 'Key value'
# pd.merge(df1,df2,on="key",how="inner")

pd.merge(df1,df2,on="key",how="outer")  #when no value match given output will NAN
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="right")
```
