### Seaborn
Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Basic plotting with seaborn
tips =sns.load_dataset('tips')
tips
```

```python
## Creating scatter plot

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter plot of total bill ")
plt.show()
```

```python
## Line plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line plot are using the")
plt.show()
```

```python
## Cretical plot
## Bar Plot

sns.barplot(x='day',y='total_bill',data=tips)
plt.title("critical plot are given as to be")
plt.show()
```

```python
## Box plot 
sns.boxenplot(x='day',y='total_bill',data=tips)
# plt.title('Box data as given as')
```

```python
## Voilen Plot

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
## Crat histogram
sns.histplot(tips['total_bill'],bins=10,kde=True)  # kde(curve ) call it, Bins = Bar lines 
```

```python
## Construct will KDE Plot

sns.kdeplot(tips['total_bill'],shade=True)
```

```python
## Pair plot

sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
## Crat heat map
cor = tips[['total_bill','tip','size']].corr()
cor
```

```python
sns.heatmap(cor,annot=True,cmap='coolwarm')  ## annot use to repersente number   
```

```python
import pandas as pd
df =pd.read_csv("sales.csv")
df.head()
```

```python
## plot total figure
plt.figure(figsize=(12,6))
sns.barplot(x='product',y='total_sales',data=df,estimator=sum)
plt.title("The bar plot are is that")
plt.xlabel("x-axis")
plt.ylabel('y-axix')
plt.show()
```
