### Pandas :

Pandas is an open-source Python library used for data analysis and manipulation. It provides fast, flexible, and easy-to-use data structures for working with structured data such as tables, spreadsheets, and databases.

```python
import pandas as pd
```

```python
# Series 

import pandas as pd
data =[1,2,3,4,5,6]
series =pd.Series(data)
print(series)
```

```python
# Create a series From dicstionary

Data ={"a":1,"b":2,"c":3}

dic_series =pd.Series(Data)
print(dic_series)
```

```python
data =([10,20,30])
index=['a','b','c']

pd.Series(data,index=index)
```
