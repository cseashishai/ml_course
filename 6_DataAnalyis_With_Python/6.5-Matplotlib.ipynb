### Data Visulization with matplotLib
Data Visualization is the graphical representation of data to help understand patterns, trends, relationships, and insights using visual elements like charts and graphs.

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[8,12,14,18]

# Now i ploting a line graph
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("That is basic line plot")
plt.show()
```

```python
x=[1,2,3,4]
y=[8,12,14,18]
 
#Creat a coustmize line plot

plt.plot(x,y, color= 'blue',linestyle=':',marker='o')  #color will be red  and linestyle '',marker'o' refer brek point
```

```python
x=[1,2,3,4]
y=[8,12,14,18]
 
#Creat a coustmize line plot

plt.plot(x,y, color= 'blue',linestyle=':',marker='o',linewidth=3,markersize =9) # it provides of line theker 
plt.grid(True) #grid provides to line of graph
```

```python
## Multiplot Plots
## Simple plot

x=[1,2,3,4]
y1=[8,12,14,18]
y2=[1,2,3,4]

plt.Figure(figsize=(8,6))  # 8-width, 6-height
plt.subplot(2,2,1)  # nrow ,#ncol, #index 1
plt.plot(x,y1,color='green')
plt.title("The ploting 1 ")

plt.subplot(2,2,2)  # nrow ,#ncol, #index 2
plt.plot(y1,x,color='green')
plt.title("The ploting 2")

plt.subplot(2,2,4)  # nrow ,#ncol, #index 4
plt.plot(x,y2,color='blue')
plt.title("The ploting 3")

```

```python
# ## Bar Plot 

categries = ['A','B','C','D','E']
value =[5,6,7,8,6] 
 
# Creat bar plot
plt.bar(categries,value,color='purple')
plt.xlabel('categries')
plt.ylabel('value')
plt.title('Bar plot')
plt.show()
```

### Histogram
A histogram is a graphical representation of numerical data that shows how data values are distributed across different intervals (called bins), using adjacent bars to represent the frequency of values in each interval.

```python
data =[1,2,3,4,6,2,3,1,4,6,4,5,4,6,5]

# crating the histogram

plt.hist(data,bins=5,color='orange',edgecolor='green') # bins is require to the devided this  perticuler data
```

```python
## Creating Sccater plot
# sample data
x=[1,2,3,4,5]
y=[2,3,4,5,6]

plt.scatter(x,y,color='blue',marker='x') #scatter showing between  two points  relestionship

```

```python
## Pi chart
labeles =['A','B','C','D']
size =[30,20,40,10]
colors =['gold','skyblue','lightcoral','yellowgreen']
explode =(0.2,0,0,0) # Explode use move out the 1st slice

# creat pi chart 

plt.pie(size,labels=labeles,colors=colors,autopct="%1.1f%%",explode=explode,shadow=True) # autopct use to get persentage values
plt.show()
```

```python
# Sales data visulazation

import pandas as pd
import matplotlib.pyplot as plt
df =pd.read_csv('sales.csv')
df
```

```python
total_sales_by_product =df.groupby('product')['unit_price'].sum()
print(total_sales_by_product)
```

```python
total_sales_by_product.plot(kind='bar',color='teal')# kind use to findig which graph to repesent
```

```python
## plot sales trends dates
sales_trand_date=df.groupby('date')['total_sales'].sum().reset_index()
plt.plot(sales_trand_date['date'],sales_trand_date['total_sales']) #it may be use to persenting max value data
```
