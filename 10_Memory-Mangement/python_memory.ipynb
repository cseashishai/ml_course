### Python Memory Management

Memory management in Python involves a combination of automatic garbage collection, reference counting, and various internal optimizations to efficiently manage memory allocation and deallocation. Understanding these mechanisms can help developers write more efficient and robust applications.

Topics Covered

Key Concepts in Python Memory Management

Memory Allocation and Deallocation

Reference Counting

Garbage Collection

The gc Module

Memory Management Best Practices

```python

import sys
import gc


# 1. Memory Allocation and Deallocation

def memory_allocation_example():
    print("\n--- Memory Allocation Example ---")
    a = [1, 2, 3, 4]
    print("List created:", a)

    b = a
    print("Another reference created")

    del a
    print("Deleted variable 'a', but list still exists because 'b' references it:", b)

    del b
    print("Deleted variable 'b' -> object can now be garbage collected")
```

```python
 ## Reference Counting

def reference_counting_example():
    print("\n--- Reference Counting Example ---")

    x = []
    print("Initial reference count:", sys.getrefcount(x))

    y = x
    print("After assigning y = x:", sys.getrefcount(x))

    z = x
    print("After assigning z = x:", sys.getrefcount(x))

    del y
    print("After deleting y:", sys.getrefcount(x))

    del z
    print("After deleting z:", sys.getrefcount(x))

```

```python
# 3. Garbage Collection (Circular Reference Example)

class Node:
    def __init__(self):
        self.ref = None


def garbage_collection_example():
    print("\n--- Garbage Collection Example (Circular Reference) ---")

    a = Node()
    b = Node()

    a.ref = b
    b.ref = a

    print("Created circular reference between two objects")

    del a
    del b

    print("Deleted references. Python GC will clean circular references.")
```

```python
# 4. Using the gc Module

def gc_module_example():
    print("\n--- gc Module Example ---")

    print("GC Enabled:", gc.isenabled())

    collected = gc.collect()
    print("Garbage collector collected", collected, "objects")

    print("Garbage collector thresholds:", gc.get_threshold())
```

```python
# 5. Memory Management Best Practices

def best_practices_example():
    print("\n--- Best Practices Example ---")

    # Using generators instead of large lists
    def number_generator(n):
        for i in range(n):
            yield i

    gen = number_generator(1000000)

    print("Generator created (memory efficient)")

    for _ in range(5):
        print(next(gen))
```
