### Python logging 
Logging in Python is the process of recording messages about a program’s execution (such as errors, warnings, or general information) so developers can monitor, debug, and track what the program is doing.

```python
import logging

# Configure the basic setting of logging
logging.basicConfig(level=logging.DEBUG)
force=True

# lOGG msg

logging.debug("This is Debug massage")
logging.info("That will provided to info")
logging.warning("give to warning msg") 
logging.error("GIven error msg")
logging.critical("this critical msg")
```

### Log Level

In Python’s logging, log levels represent the severity or importance of a message. They help control which messages are recorded or displayed.

| Level        | Numeric Value | Meaning                                               |
| ------------ | ------------- | ----------------------------------------------------- |
| **DEBUG**    | 10            | Detailed information used for debugging               |
| **INFO**     | 20            | General information about program execution           |
| **WARNING**  | 30            | Something unexpected happened, but program still runs |
| **ERROR**    | 40            | A serious issue occurred                              |
| **CRITICAL** | 50            | Very severe error that may stop the program           |


```python
## Configuring Logging
import logging
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S:'
)

# lOGG msg

logging.debug("This is Debug massage")
logging.info("That will provided to info")
logging.warning("give to warning msg") 
logging.error("GIven error msg")
logging.critical("this critical msg")
```

Some of More Likes 
| Specifier | Meaning              | Example                |
| --------- | -------------------- | ---------------------- |
| `%s`      | String               | `"Hello %s" % "Ashu"`  |
| `%d`      | Integer              | `"Age %d" % 21`        |
| `%f`      | Float                | `"Price %f" % 10.5`    |
| `%0.2f`   | Float with 2 decimal | `"Value %.2f" % 3.456` |
| `%x`      | Hexadecimal          | `"Hex %x" % 255`       |
| `%o`      | Octal                | `"Octal %o" % 10       |
| `%e`      | Scientific notation  | `"Sci %e" % 1000`      |
