```python
import logging

#Creat logger for module
loger1 =logging.getLogger("module1")
loger1.setLevel(logging.DEBUG)

#creat logger module2

loger2=logging.getLogger("Module2")
loger2.setLevel(logging.WARNING)

#Creating Logger setting
logging.basicConfig(  
    level=logging.DEBUG,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', # its br creat date time or such or more such of directory
    datefmt='%Y-%m-%d %H:%M:%S' # its be creating year , month,hourse,minute
    # force=True
)
```

```python
## log msg are diffrent Loger
loger1.debug("This module is debugs")
loger2.warning("this is a warning msg")
loger2.error("Its may be provides error")
```
